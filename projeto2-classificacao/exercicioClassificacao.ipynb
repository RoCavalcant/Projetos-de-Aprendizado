```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split

```

```python
# Carregando o dataset Iris
iris = load_iris()

data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['target'] = iris.target
data.head()
```

```python
data.columns
```

```python
data = data.rename(columns={'sepal length (cm)': 'comprimento_sepala',
                            'sepal width (cm)': 'largura_sepala',
                            'petal length (cm)': 'comprimento_petala',
                            'petal width (cm)': 'largura_petala'})
data.tail()
```

```python
x = data.drop('target', axis=1)
y = data['target']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=200)
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo = DecisionTreeClassifier(max_depth=3)
modelo.fit(x_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score # Import the accuracy_score function

modelo = DecisionTreeClassifier()
modelo.fit(x_train, y_train)

acuracia = accuracy_score(y_test, modelo.predict(x_test)) # Call accuracy_score correctly
print(f'Acurácia: {acuracia}')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# Seu modelo de árvore de decisão
modelo = DecisionTreeClassifier(max_depth=3, min_samples_split=5, min_samples_leaf=4)

# Aplicando validação cruzada com 5 folds
scores = cross_val_score(modelo, x_train, y_train, cv=5)

# Exibindo os resultados
print(f'Validação cruzada - Acurácias por fold: {scores}')
print(f'Acurácia média: {scores.mean()}')
print(f'Desvio padrão: {scores.std()}')
```

```python
# Criando uma base de dados fictícia com características de flores
novos_dados = pd.DataFrame({
    'comprimento_sepala': [5.1, 6.5, 4.9, 5.9, 6.7],
    'largura_sepala': [3.5, 3.0, 3.1, 3.2, 3.3],
    'comprimento_petala': [1.4, 5.8, 1.5, 4.8, 5.7],
    'largura_petala': [0.2, 2.2, 0.1, 1.8, 2.1]
})
novos_dados.head()
```

```python
modelo_ajustado = DecisionTreeClassifier(max_depth=3, min_samples_split=5, min_samples_leaf=4)

modelo_ajustado.fit(x_train, y_train)  # x_train e y_train são seus dados de treino

previsoes = modelo_ajustado.predict(novos_dados)

# Exibindo as previsões
print(f'Previsões para os novos dados: {previsoes}')
```

```python
# Transformando rótulos numéricos nas classes reais
nomes_classes = ['Setosa', 'Versicolor', 'Virginica']
previsoes_nomeadas = [nomes_classes[pred] for pred in previsoes]

print(f'Previsões (nomes das classes): {previsoes_nomeadas}')
```
